```python
import pandas as pd
pd.set_option('display.max_rows', 500)
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('groceries.csv')
```

```python
df.head(3)
```

```python
df.shape
```

```python
market = []
for i in range (0, df.shape[0]):
    cus = []
    for j in df.columns:
        if type(df[j][i])==str:
            cus.append(df[j][i])
            
    market.append(cus)           
```

```python
market
```

```python
l = []
for i in market:
    for j in i:
        l.append(j)
```

```python
l
```

```python
import collections
```

```python
freq = collections.Counter(l)
```

```python
freq.keys()
```

```python
freq.values()
```

```python
d = {'Item Name' : freq.keys(),
    'Values' : freq.values()}
```

```python
pd.DataFrame(d).sort_values(by = 'Values', ascending = False)
```
